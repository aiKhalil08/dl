```python
from utils import w_sum

def neural_network(input, weights):
    pred = vec_mat_mult(input, weights)
    return pred

def vec_mat_mult(vec, mat):
    assert(len(vec) == len(mat))

    output = [0,0,0]

    for i in range(len(vec)):
        output[i] = w_sum(vec, mat[i])

    return output
```

```python
weights = [
    [0.1, 0.1, -0.3], # output 1
    [0.1, 0.2, 0.0], # output 2
    [0.0, 1.3, 0.1], # output 3
]

# weights = [
#     [0.5, -0.1, -0.4], # output 1
#     [0.9, 0.6, -0.2], # output 2
#     [1.7, -0.3, 0.0], # output 3
# ]

toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65,0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

hurt = [0.1, 0.0, 0.0, 0.1] # true 1
win = [1, 1, 0, 1] # true 2
sad = [0.1, 0.0, 0.1, 0.2] # true 3

alpha = 0.01
input = [toes[0],wlrec[0],nfans[0]]
true = [hurt[0], win[0], sad[0]]
```

```python
for iter in range(10):
    print(f"iteration {iter + 1}")
    print(f"weights = {weights}")
    # 1. predict
    pred = neural_network(input, weights)
    print(f"prediction = {pred}, true = {true}")

    # 2. compare
    errors = [0,0,0]
    deltas = [0,0,0]

    for i in range(len(pred)):
        errors[i] = (pred[i] - true[i]) ** 2
        deltas[i] = pred[i] - true[i]
    print(f"errors = {errors}, deltas = {deltas}")

    # 3. adjust
    weight_deltas = [
        [0,0,0], # for error #1
        [0,0,0], # for error #2
        [0,0,0], # for error #3
    ]

    for j in range(len(weights)):
        for k in range(len(input)):
            weight_deltas[j][k] = input[k] * deltas[j]
    print(f"weight_deltas = {weight_deltas}\n")

    for l in range(len(weights)):
        for m in range(len(input)):
            weights[l][m] -= alpha * weight_deltas[l][m]
```
